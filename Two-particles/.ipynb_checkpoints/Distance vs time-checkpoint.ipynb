# Distance vs Time

In this Script, I will plot the distance among two particles in a channel vs the time. The particles are moving along the channel due to a flow field. 

```python
Q1 = pd.read_csv('distance-interaction_Q1ul-min.dat', sep = '\t', index_col = 0)
Q2 = pd.read_csv('distance-interaction_Q2ul-min.dat', sep = '\t', index_col = 0)
Q2_5 = pd.read_csv('distance-interaction_Q2,5ul-min.dat', sep = '\t', index_col = 0)
Q4 = pd.read_csv('distance-interaction_Q4ul-min.dat', sep = '\t', index_col = 0)

Q1n = pd.read_csv('distance-Nointeraction_Q1ul-min.dat', sep = '\t', index_col = 0)
Q2n = pd.read_csv('distance-Nointeraction_Q2ul-min.dat', sep = '\t', index_col = 0)
Q2_5n = pd.read_csv('distance-Nointeraction_Q2,5ul-min.dat', sep = '\t', index_col = 0)
Q4n = pd.read_csv('distance-Nointeraction_Q4ul-min.dat', sep = '\t', index_col = 0)
```

```python
plt.plot(Q1['t_inter(s)'], Q1['d_inter(um)'], 'o', label = 'Q = 1 ml/min', c ='blue')
plt.plot(Q2['t_inter(s)'], Q2['d_inter(um)'], 'o', label = 'Q = 2 ml/min', c ='red')
plt.plot(Q2_5['t_inter(s)'], Q2_5['d_inter(um)'], 'o', label = 'Q = 2,5 ml/min', c= 'orange')
plt.plot(Q4['t_inter(s)'], Q4['d_inter(um)'], 'o', label = 'Q = 4 ml/min', c = 'pink')

plt.plot(Q1n['t_Ninter(s)'], Q1n['d_Ninter(um)'], 'o' ,c ='blue')
plt.plot(Q2n['t_Ninter(s)'], Q2n['d_Ninter(um)'], 'o' ,c ='red')
plt.plot(Q2_5n['t_Ninter(s)'], Q2_5n['d_Ninter(um)'], 'o', c= 'orange')
plt.plot(Q4n['t_Ninter(s)'], Q4n['d_Ninter(um)'], 'o', c = 'pink')


plt.ylabel(r'$distance (\mu m)$', size = 40)
plt.xlabel('time (s)', size = 40)

plt.legend(fontsize = 16, bbox_to_anchor=(1.7, 1))
```
