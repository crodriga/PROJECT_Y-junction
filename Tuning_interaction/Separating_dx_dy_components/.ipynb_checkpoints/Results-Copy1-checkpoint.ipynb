# Mean

```python
def mean_dy_and_mag(dataframe, angle = int):
    
    dataframe['dy_conv'] =  (dataframe['dy'])/18.82
    
    exp = []
    
    exp.append(dataframe.query('dy_conv < 0.025').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.025 & dy_conv < 0.05').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.05 & dy_conv < 0.075').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.075 & dy_conv < 0.1').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.1 & dy_conv < 0.125').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.125 & dy_conv < 0.15').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.15 & dy_conv < 0.175').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.175 & dy_conv < 0.2').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.2 & dy_conv < 0.225').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.225 & dy_conv < 0.25').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.25 & dy_conv < 0.275').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.275 & dy_conv < 0.3').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.3 & dy_conv < 0.325').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.325 & dy_conv < 0.35').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.35 & dy_conv < 0.375').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.4 & dy_conv < 0.425').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.425 & dy_conv < 0.45').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.45 & dy_conv < 0.475').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.475 & dy_conv < 0.5').mean(axis = 0).to_frame())
    exp.append(dataframe.query('dy_conv > 0.5').mean(axis = 0).to_frame())
    
    m = []
    d = []
    for i in range(0, len(exp)):

        m.append(exp[i].iloc[6].values) # magnetization
        d.append(exp[i].iloc[13].values) # dy_conv
        
    mag = np.concatenate(m).ravel()
    dist = np.concatenate(d).ravel()
    
    d = {'magnetization': mag, 'dy': dist, 'angle': angle}

    df = pd.DataFrame(data=d)
    
    return df
```
