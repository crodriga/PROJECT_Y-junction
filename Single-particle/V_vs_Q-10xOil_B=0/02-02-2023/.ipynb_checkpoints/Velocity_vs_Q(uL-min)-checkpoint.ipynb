## Computing particle velocity

```python
def v(df):   
    
    delta_r = [0]

    for i in range(0,len(df)-1):
        j = i+1  
        delta_r.append(np.sqrt((df.iloc[j].x-df.iloc[i].x)**2+(df.iloc[j].y-df.iloc[i].y)**2))
        
    df['delta_r'] = delta_r
    df['delta_t'] = 0.04
    df['v'] = df['delta_r']/df['delta_t']
    
    df.drop(df[df['v'] >= 200].index, inplace = True)
    return df
```

```python
%%time
d_one = delta_r(one)
d_two = delta_r(two)
d_three = delta_r(three)
d_one_half = delta_r(one_half)
d_two_half = delta_r(two_half)
```
