# Reading trajectories
In this script, I will read the data generated by the program ImageJ and in particular the plugging TrackMate.

```python
directory = "/media/Carol/Magnetic/Experiments/Microfluidics_experiments/PDMS_chips/16-02-2023/Pre-treated-videos/2ul-min"

filenames = [file for file in glob.glob(
    os.path.join(directory,('*.csv')))]
filenames
```

```python
trj0 = pd.read_csv(filenames[0], sep =",",skiprows = [1,2,3], index_col = [0], usecols = [2,4,5,8])
trj1 = pd.read_csv(filenames[1], sep =",",skiprows = [1,2,3], index_col = [0], usecols = [2,4,5,8])
trj2 = pd.read_csv(filenames[2], sep =",",skiprows = [1,2,3], index_col = [0], usecols = [2,4,5,8])
trj3 = pd.read_csv(filenames[3], sep =",",skiprows = [1,2,3], index_col = [0], usecols = [2,4,5,8])
trj4 = pd.read_csv(filenames[4], sep =",",skiprows = [1,2,3], index_col = [0], usecols = [2,4,5,8])

trj4.index.get_level_values('TRACK_ID').unique()
```

```python
p0 = trj0.query('TRACK_ID == 2').sort_values(by = ['FRAME'])
p1 = trj1.query('TRACK_ID == 0').sort_values(by = ['FRAME'])
p2 = trj1.query('TRACK_ID == 4').sort_values(by = ['FRAME'])
p3 = trj1.query('TRACK_ID == 39').sort_values(by = ['FRAME'])
p4 = trj2.query('TRACK_ID == 0').sort_values(by = ['FRAME'])
p5 = trj3.query('TRACK_ID == 0').sort_values(by = ['FRAME'])
p6 = trj4.query('TRACK_ID == 3').sort_values(by = ['FRAME'])
```

```python
plt.plot(p0.POSITION_Y, p0.POSITION_X, '.')
plt.plot(p1.POSITION_Y, p1.POSITION_X, '.')
plt.plot(p2.POSITION_Y, p2.POSITION_X, '.')
plt.plot(p3.POSITION_Y, p3.POSITION_X, '.')

entrance= 375
bifurcation = 140
#plt.xlim(100,200)

plt.axvline(x=entrance)
plt.axvline(x=bifurcation)
```

```python
def v(df,frame_rate = float , pix = float, um = float):   
    
    delta_r = [0]

    for i in range(0,len(df)-1):
        j = i+1  
        delta_r.append(np.sqrt((df.iloc[j].POSITION_X-df.iloc[i].POSITION_X)**2+(df.iloc[j].POSITION_Y-df.iloc[i].POSITION_Y)**2))
    
    df['y'] = df['POSITION_Y']*(um/pix)  # um/pix from scale bar
    df['delta_r'] = delta_r 
    df['delta_t'] = (frame_rate) # 30 fps
    df['v'] = (df['delta_r']*(um/pix))/df['delta_t'] # um/pix from scale bar
    
    return df
```

```python
v(p0, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p1, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p2, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p3, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p4, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p5, frame_rate = 1/30 , pix = 497.549, um = 1000)
v(p6, frame_rate = 1/30 , pix = 497.549, um = 1000)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))

plt.plot(p0.y, p0.v, '.')
plt.plot(p1.y, p1.v, '.')
plt.plot(p2.y, p2.v, '.')
plt.plot(p3.y, p3.v, '.', c = 'green')

plt.xlabel(r'$y (\mu m)$')
plt.ylabel(r'$v( \mu m /s)$')

#plt.ylim(0,1)
```

```python
def channel_v_vs_position(df_particle, entrance = int, bifurcation = int):
    
    string = ('y <= '+str(entrance)+' & y > '+str(bifurcation)+' ')
    
    entrance = df_particle.query('y > '+str(entrance))
    channel = df_particle.query(string)
    bifurcation = df_particle.query('y < '+str(bifurcation))
    
    v_entrance = entrance.loc[:, 'v'].mean()
    v_channel = channel.loc[:, 'v'].mean()
    v_bifurcation = bifurcation.loc[:, 'v'].mean()
    
    return v_entrance, v_channel, v_bifurcation
```

```python
entrance= 375*1000/497.549
bifurcation = 140*1000/497.549
```

```python
ve0, vc0, vb0 = channel_v_vs_position(p0, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve1, vc1, vb1 = channel_v_vs_position(p1, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve2, vc2, vb2 = channel_v_vs_position(p2, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve3, vc3, vb3 = channel_v_vs_position(p3, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve4, vc4, vb4 = channel_v_vs_position(p4, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve5, vc5, vb5 = channel_v_vs_position(p4, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
ve6, vc6, vb6 = channel_v_vs_position(p4, entrance = 375*1000/497.549, bifurcation =140*1000/497.549)
```
