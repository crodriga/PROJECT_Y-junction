```python
def v(df,frame_rate = float , pix = float, um = float):   
    
    delta_r = [0]

    for i in range(0,len(df)-1):
        j = i+1  
        delta_r.append(np.sqrt((df.iloc[j].POSITION_X-df.iloc[i].POSITION_X)**2+(df.iloc[j].POSITION_Y-df.iloc[i].POSITION_Y)**2))
    
    df['y'] = df['POSITION_Y']*(um/pix)  # um/pix from scale bar
    df['delta_r'] = delta_r 
    df['delta_t'] = (frame_rate) # 30 fps
    df['v'] = (df['delta_r']*(um/pix))/df['delta_t'] # um/pix from scale bar
    
    return df
```

```python
def channel_v_vs_position(df_particle, entrance_value = float, bifurcation_value = float):
    
    string_entrance = ('y > '+str(entrance_value))
    string_channel = ('y <= '+str(entrance_value)+' & y > '+str(bifurcation_value)+' ')
    string_bifurcation = ('y < '+str(bifurcation_value))
    
    entrance = df_particle.query(string_entrance)
    channel = df_particle.query(string_channel)
    bifurcation = df_particle.query(string_bifurcation)

    v_entrance = entrance.loc[:, 'v'].mean()
    v_channel = channel.loc[:, 'v'].mean()
    v_bifurcation = bifurcation.loc[:, 'v'].mean()
    
    return v_entrance, v_channel, v_bifurcation 
```

```python
entrance= 375*1000/497.549
bifurcation = 140*1000/497.549
```

```python
ve0, vc0, vb0 = channel_v_vs_position(p0, entrance_value = entrance, bifurcation_value =bifurcation)
ve1, vc1, vb1 = channel_v_vs_position(p1, entrance_value = entrance, bifurcation_value =bifurcation)
ve2, vc2, vb2 = channel_v_vs_position(p2, entrance_value = entrance, bifurcation_value =bifurcation)
ve3, vc3, vb3 = channel_v_vs_position(p3, entrance_value = entrance, bifurcation_value =bifurcation)
ve4, vc4, vb4 = channel_v_vs_position(p4, entrance_value = entrance, bifurcation_value =bifurcation)
```
